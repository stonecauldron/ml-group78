```python
# Useful starting lines
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
%load_ext autoreload
%autoreload 2
```

# Data Exploration

```python
# load the data
from proj1_helpers import load_csv_data
DATA_TRAIN_PATH = '../train.csv'
y, tX, ids = load_csv_data(DATA_TRAIN_PATH)

# useful globals
N = tX.shape[0]
D = tX.shape[1]
```

## Number of invalid values per feature

```python
def nb_invalid_data(col_x, invalid_value=-999):
    """Counts the number of invalid data points in a feature"""
    col = tX[:, 0]
    return col_x[col_x == invalid_value].shape[0]
```

```python
nb_invalid_values = np.zeros(D)
for i in range(D):
    nb_invalid_values[i] = nb_invalid_data(tX[:,i])
    
plt.bar(range(30), nb_invalid_values, width=0.5)

plt.ylim([0, N])
plt.title("Number of invalid values")
plt.xlabel("feature index")
plt.ylabel("total number")
plt.minorticks_on()
plt.grid(True)
plt.show()
```

## Feature distribution

```python
def clean_up_invalid_values(col_x, invalid_value=-999):
    col_x
    mean = col_x[col_x != -999].mean()
    col_x[col_x == -999] = mean
    return col_x

def feature_hist(col_x, bins=50, title=''):
    plt.hist(col_x, bins=bins)
    
    plt.title(title)
    plt.ylim([0, N])
    plt.grid(True)
```

```python
# plot the feature distribution
nb_cols_subplot = 2
nb_rows_subplot = np.ceil(tX.shape[1]/nb_cols_subplot)

plt.figure(figsize=(10,50))
for i in range(tX.shape[1]):
    plt.subplot(nb_rows_subplot, nb_cols_subplot, i + 1)
    feature_hist(tX[:,i], bins=50, title='feature ' + str(i + 1))
plt.show()
```
